```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
!unzip -o /content/census+income.zip
df = pd.read_csv('/content/adult.data')
```

```python
df.head()
```

```python
# Nama kolom (standar Adult Dataset)
columns = [
    'age', 'workclass', 'fnlwgt', 'education', 'education_num',
    'marital_status', 'occupation', 'relationship', 'race',
    'sex', 'capital_gain', 'capital_loss', 'hours_per_week',
    'native_country', 'income'
]

df.columns = columns

df.head()
```

```python
df.info()
```

```python
df.tail()
```

```python
df.describe()
```

```python
df.replace(' ?', np.nan, inplace=True)
```

```python
df.isnull().sum()
```

```python
df.duplicated().sum()
```

```python
df = df.drop_duplicates()
```

```python
num_cols = df.select_dtypes(include=['int64', 'float64']).columns

for col in num_cols:
    plt.figure(figsize=(6,3))
    sns.boxplot(x=df[col])
    plt.title(f'Boxplot {col}')
    plt.show()
```

```python
df['income'].value_counts()
```

```python
sns.countplot(x='income', data=df, palette='viridis')
plt.title('Distribusi Income')
plt.show()
```

```python
num_cols = df.select_dtypes(include=['int64', 'float64']).columns
cat_cols = df.select_dtypes(include=['object']).columns

print("Numerik:", num_cols)
print("Kategorik:", cat_cols)
```

```python
for col in num_cols:
    plt.figure(figsize=(6,3))
    sns.histplot(df[col], kde=True)
    plt.title(f'Distribusi {col}')
    plt.show()
```

```python
for col in cat_cols:
    plt.figure(figsize=(20,8))
    sns.countplot(x=df[col])
    plt.title(f'Distribusi {col}')
    plt.xticks(rotation=90)
    plt.show()
```

```python
for col in cat_cols:
    df[col] = df[col].str.strip()
```
